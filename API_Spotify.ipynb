```python
# Token de acceso (reemplázalo con uno válido)
access_token = access_token

# ID del artista (cámbialo por el artista que quieras)
artist_id = "6EMoWayH9WulC8F8b6m3BH"

# URLs de la API
artist_url = f"https://api.spotify.com/v1/artists/{artist_id}"
albums_url = f"https://api.spotify.com/v1/artists/{artist_id}/albums?limit=50&include_groups=album,single"

# Encabezados de autorización
headers = {
    "Authorization": f"Bearer {access_token}"
}

# 1. Obtener información del artista
response = requests.get(artist_url, headers=headers)
if response.status_code == 200:
    artist_data = response.json()
    artist_info = {
        "artist_name": artist_data['name'],
        "popularity": artist_data['popularity'],
        "genres": ", ".join(artist_data['genres']),
        "followers": artist_data['followers']['total'],
        "images": artist_data['images'][0]['url'] if artist_data['images'] else 'No disponible',
        "Spotify_artist": artist_data['external_urls']['spotify']
    }
else:
    print(f"❌ Error al obtener el artista: {response.status_code}")
    exit()

# 2. Obtener los álbumes del artista
response = requests.get(albums_url, headers=headers)
albums_list = []

if response.status_code == 200:
    albums_data = response.json()["items"]

    for album in albums_data:
        albums_list.append({
            "ID_Album": album["id"],
            "album_name": album["name"],
            "release_date": album["release_date"],
            "total_tracks": album["total_tracks"],
            "Spotify_Album": album["external_urls"]["spotify"]
        })

else:
    print(f"Error al obtener álbumes: {response.status_code}")

# 3. Obtener las canciones de cada álbum
tracks_list = []

for album in albums_list:
    album_id = album["ID_Album"]
    tracks_url = f"https://api.spotify.com/v1/albums/{album_id}/tracks?limit=50"
    response_tracks = requests.get(tracks_url, headers=headers)

    if response_tracks.status_code == 200:
        tracks_data = response_tracks.json()["items"]

        for track in tracks_data:
            track_id = track["id"]
            track_name = track["name"]
            track_duration = track["duration_ms"] // 1000  # Convertir ms a segundos
            track_url = track["external_urls"]["spotify"]

            # Obtener popularidad de la canción
            track_details_url = f"https://api.spotify.com/v1/tracks/{track_id}"
            response_track_details = requests.get(track_details_url, headers=headers)

            if response_track_details.status_code == 200:
                track_details = response_track_details.json()
                track_popularity = track_details["popularity"]  # Popularidad ≈ Reproducciones
            else:
                track_popularity = "Unkwown"

            tracks_list.append({
                "album_id": album_id,
                "song_name": track_name,
                "track_duration_secs": track_duration,
                "track_popularity": track_popularity,
                "track_url": track_url
            })
    
    time.sleep(0.5)  # Espera para evitar bloqueos de la API

# 4. Convertir a DataFrames
df_artista = pd.DataFrame([artist_info])
df_albumes = pd.DataFrame(albums_list)
df_canciones = pd.DataFrame(tracks_list)

# 5. Hacer JOIN entre artista, álbumes y canciones (CROSS JOIN)
df_merged = df_albumes.merge(df_canciones, on="ID_Album", how="left")  # JOIN de álbumes con canciones
df_final = df_merged.assign(**df_artista.iloc[0])  # CROSS JOIN con la info del artista

# 6. Guardar a CSV
nombre_archivo = "Spotify_Datos.csv"
df_final.to_csv(nombre_archivo, index=False)
print(f" Datos guardados en '{nombre_archivo}' correctamente.")

# 7. Mostrar los primeros registros
print(df_final.head())
```
