```python
#Function to read, open and parse with bs4
def read_html(name):
    with open(name, encoding='utf-8') as archivo:
        contenido = archivo.read()
    return BeautifulSoup(contenido, 'html.parser')

soup = read_html(file_path)
```

```python
rest_names = []
n = 36
for name in soup.find_all('a', class_= 'BMQDV _F Gv wSSLS SwZTJ FGwzt ukgoS'):
  rest_names.append(name.text.strip())
```

```python
#Search for specific content and storage it in a list
cost_range = []
n = 36
for name in soup.find_all('span', class_= 'YECgr', limit = n): #in 'div',{'class':'listing_title'} you will have to add the path in the html code depending on the content you need to extract
  cost_range.append(name.text.strip())
```

```python
reviews = []
n = 36
for name in soup.find_all('span', class_ = 'IiChw', limit = n):
  reviews.append(name.text.strip())
```

```python
bubbles = []
type_of_food = []
price_range = []

for item in cost_range:
    if "bubbles" in item:
        bubbles.append(item.split('\n')[0])  # Toma la primera línea
    elif item.isalpha():  # Verifica si son solo letras (tipo de comida)
        type_of_food.append(item)
    elif '$' in item:  # Verifica si contiene el símbolo de dólar (rango de precios)
        price_range.append(item)
```
